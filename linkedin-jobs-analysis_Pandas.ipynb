```python
# Filter the DataFrame to include rows where the company column has null values
null_company_links = null_company_df["job_link"]

# Print all the job_link values
for job_link in null_company_links:
    print(job_link)
```

```python
#Getting the name from the link
import re

# Define a function to extract company name from job link
def extract_company_name(job_link):
    match = re.search(r'(?<=-at-)(.*?)(?=-\d+$)', job_link)
    if match:
        return match.group(0)
    else:
        return None

# Update the 'company' column where it is null based on 'job_link'
null_company_mask = df_pandas_postings['company'].isnull()
df_pandas_postings.loc[null_company_mask, 'company'] = df_pandas_postings.loc[null_company_mask, 'job_link'].apply(extract_company_name)

# Show the updated DataFrame
print(df_pandas_postings)
```

```python
# Compute null counts for each column
null_counts = df_pandas_postings.isnull().sum()

# Rename the resulting series to "Null Counts"
null_counts = null_counts.rename("Null Counts")

# Display the null counts
print(null_counts)
```

```python
# Compute null counts for each column
null_counts = df_pandas_skills.isnull().sum()

# Rename the resulting series to "Null Counts"
null_counts = null_counts.rename("Null Counts")

# Display the null counts
print(null_counts)
```

```python
# Count the number of rows in the DataFrame
row_count = df_pandas_skills.shape[0]

# Print the number of rows
print("Number of rows in df_pandas_skills:", row_count)
```

```python
# Drop rows where job_skills is NULL
df_pandas_skills = df_pandas_skills.dropna(subset=["job_skills"])

# Check the number of rows after dropping NULL values
row_count_cleaned = df_pandas_skills.shape[0]

# Print the number of rows after dropping NULL values
print("Number of rows in df_pandas_skills after dropping NULL values in job_skills:", row_count_cleaned)
```

```python
# Concatenate all job skills into a single string separated by ','
skills_concatenated = df_pandas_skills["job_skills"].str.cat(sep=', ')

# Convert the concatenated string to lowercase
skills_lower = skills_concatenated.lower()

# Split the concatenated string into a list of skills
skills_list = skills_lower.split(", ")

# Show the first 10 skills and the total count
skills_list[:10], len(skills_list)
```

```python
# Split the concatenated string into separate skills
skills_list = skills_lower.split(", ")

# Remove the words "skills" or "skill" from the skill names
skills_list_cleaned = [skill.strip().replace('skills', '').replace('skill', '') for skill in skills_list]

# Convert the list to a Pandas DataFrame
exploded_skills = pd.DataFrame({"skill": skills_list_cleaned})

# Group by skill and count the occurrences of each skill
skill_counts = exploded_skills["skill"].value_counts().reset_index()

# Rename the columns
skill_counts.columns = ['skill', 'count']

# Order the DataFrame by count in descending order
skill_counts_descending = skill_counts.sort_values(by='count', ascending=False)

# Show the resulting DataFrame
print(skill_counts_descending.head(30))
```

```python
import matplotlib.pyplot as plt

# Get the top 30 skills and their counts
top_skills = skill_counts_descending.head(30)

# Create a bar plot using Matplotlib
plt.figure(figsize=(12, 8))
plt.bar(top_skills['skill'], top_skills['count'], color='skyblue')
plt.title('Top 30 Skills')
plt.xlabel('Skills')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Get the top 30 skills and their counts
top_skills = skill_counts_descending.head(30)

# Create a dictionary from the skill counts data
wordcloud_data = dict(zip(top_skills['skill'], top_skills['count']))

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(wordcloud_data)

# Plot the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Top 30 Skills Word Cloud')
plt.show()
```
